```python
Datadirectory = r"C:\Users\bened\Downloads\DL(Drowsiness detection)\archive\train" 
Classes = ["yawn","no_yawn"]##List of Classes
for category in Classes:
    path = os.path.join(Datadirectory, category) ## //
    for img in os.listdir(path):
        img_array1 = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        backtorgb = cv2.cvtColor(img_array1,cv2.COLOR_GRAY2RGB)
        plt.imshow(img_array1,cmap="gray")
        plt.show()
        break
    break
```
