```python
import tensorflow as tf
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
Datadirectory = "C:/Users/bened/Downloads/Real-time Drowsiness Detection/train(celebrity)" ##training dataset
Classes = ["Angelina_Jolie","Brad_Pitt"] ##List of Classes
```

```python
training_Data=[]

def create_training_Data():
    for category in Classes:
        path =os.path.join(Datadirectory, category)
        class_num = Classes.index(category) ## 0,1
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
                backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2RGB) 
                new_array = cv2.resize(backtorgb, (img_size,img_size))
                training_Data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

```python
import random

random.shuffle(training_Data) #shuffling to avoid overfitting
```

```python
img_size =224

X = []
y = []

for features,label in training_Data:
    X.append(features) #data
    y.append(label)

X = np.array(X).reshape(-1 , img_size, img_size, 3) #3-number of channels
```

```python
X.shape # 1448 number of images in dataset
```

```python
#normalize the data
X= X/255.0; ## we are normalizing it  
#better to use sk -learn
```

```python
Y = np.array(y) #saving labels to np array
```

```python
import pickle

# Assuming X and y are numpy arrays that you want to save
# X and y should be defined and initialized before this code snippet

# Save X using pickle
with open("X.pickle", "wb") as pickle_out:
    pickle.dump(X, pickle_out)

# Save y using pickle
with open("y.pickle", "wb") as pickle_out:
    pickle.dump(y, pickle_out)

""""
Persistence: Pickling allows you to save complex Python objects 
to disk, databases, or network storage, 
preserving their structure and state.
This is useful for saving program state, 
caching, or sharing data between different applications.
"""
```

```python
pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle","rb")
y = pickle.load(pickle_in)

#normally things get saved in ram after restarting we have to run from start
#to load the variable functionality of picker 
```

```python
import tensorflow as tf
from tensorflow import keras # tesorflow contain many predefined deeplearning algorithms
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.mobilenet.MobileNet()
```

```python
model.summary()
```

```python
base_input = model.layers[0].input ##input
```

```python
base_output = model.layers[-4].output #output - dropout layer
```

```python
Flat_layer= layers.Flatten()(base_output)
final_output = layers.Dense(1)(Flat_layer) ## one node (1/0)
final_output = layers.Activation('sigmoid')(final_output)
```

```python
new_model = keras.Model(inputs =base_input, outputs = final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
new_model.fit(X,Y,epochs =10,validation_split =0.1) ##trainig
```

```python
new_model.save('celebrityepochs10.h5')
```

```python
import winsound
import numpy as np
import cv2
import tensorflow as tf

# Load the pre-trained face and eye cascade classifiers
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt.xml')
eyeCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Load the pre-trained models for eye state detection for Angelina Jolie and Brad Pitt
new_model1 = tf.keras.models.load_model('celebrityepochs10.h5')
new_model2 = tf.keras.models.load_model('SubmitEpochs10.h5')

# Function to detect face and perform eye state detection
def detect_eye_state(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        face_roi = frame[y:y+h, x:x+w]

        eyes = eyeCascade.detectMultiScale(gray[y:y+h, x:x+w], 1.1, 4)
        for (ex, ey, ew, eh) in eyes:
            eye_roi = face_roi[ey:ey+eh, ex:ex+ew]

            final_image = cv2.resize(eye_roi, (224, 224))
            final_image = np.expand_dims(final_image, axis=0)  # Add fourth dimension
            final_image = final_image / 255.0

            celeb_prediction = new_model1.predict(final_image)
            eye_prediction = new_model2.predict(final_image)

            celeb_confidence = celeb_prediction[0][0]
            eye_confidence = eye_prediction[0][0]

            if celeb_confidence < 0.5:
                face_label = "Angelina Jolie"
            else:
                face_label = "Brad Pitt"

            # Draw rectangle around the face
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

            # Display the celebrity label on the frame
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(frame, face_label, (x, y - 10), font, 1, (0, 0, 255), 2, cv2.LINE_AA)

            # Draw rectangle around the eye
            cv2.rectangle(frame, (x+ex, y+ey), (x+ex+ew, y+ey+eh), (0, 255, 0), 2)

            # Perform eye state detection based on confidence scores
            if eye_confidence < 0.5:
                eye_status = "Closed Eyes"
            else:
                eye_status = "Open Eyes"

            # Display the eye status on the frame
            cv2.putText(frame, eye_status, (x+ex, y+ey-10), font, 0.7, (0, 0, 255), 2, cv2.LINE_AA)

    return frame

# Path to the image file
image_path = 'facescombinations.jpg'

# Read the image
image = cv2.imread(image_path)

# Perform eye state and face detection on the image
processed_image = detect_eye_state(image)

# Display the processed image
cv2.imshow('Drowsiness Detection - Photo', processed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
